```python
core_weather["target"] = core_weather.shift(-1)["temp_max"]
```

```python
core_weather
```

```python
core_weather = core_weather.iloc[:-1,:].copy()
core_weather
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
predictors = ["precip","temp_max","temp_min"]
```

```python
train = core_weather.loc[:"2020-01-01"]
```

```python
test =core_weather.loc["2020-01-02":]
```

```python
reg.fit(train[predictors],train["target"])
```

```python
predications =reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(test["target"],predications )
```

```python
combined = pd.concat([test["target"],pd.Series(predications,index=test.index)],axis=1)
combined.columns=["actual","predications"]
combined
```

```python
combined.plot()
```

```python
reg.coef_
```

```python
def create_predications(predictors,core_weather,reg):
  train = core_weather.loc[:"2020-01-01"]
  test =core_weather.loc["2020-01-02":]
  reg.fit(train[predictors],train["target"])
  predications =reg.predict(test[predictors])
  error = mean_absolute_error(test["target"],predications )
  combined = pd.concat([test["target"],pd.Series(predications,index=test.index)],axis=1)
  combined.columns=["actual","predications"]
  return error,combined
```

```python
core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean()
core_weather
```

```python
core_weather = core_weather.iloc[30:,:].copy()
```

```python
core_weather["month_day_max"] = core_weather["month_max"]/core_weather["temp_max"]
```

```python
core_weather["max_min"] = core_weather["temp_max"]/core_weather["temp_min"]
```

```python
predictors = ["precip","temp_max","temp_min","month_max","month_day_max","max_min"]
```

```python
error, combined = create_predications(predictors,core_weather,reg)
```
